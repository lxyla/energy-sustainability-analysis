## Missing Data Check
Missing values were inspected using `isna()`.

A small number of missing observations were found in very recent years too.
However thsi project will be focusing on the years 1990-2022

```python
germany_renewables.isna().sum()
```

```python
years = [str(year) for year in range(1990, 2023)]

germany_co2_clean = germany_co2[
    ["Country Name", "Country Code"] + years
]

germany_renewables_clean = germany_renewables[
    ["Country Name", "Country Code"] + years
]
```

```python
germany_co2_clean.isna().sum()
```

```python
germany_renewables_clean.isna().sum()
```

Before reshaping, identifier columns are explicitly selected and renamed to standardized names to simplify downstream analysis and merging.

```python
id_vars = ["Country Name", "Country Code"]
year_cols = [str(year) for year in range(1990, 2023)]
```

```python
germany_co2_clean = germany_co2[id_vars + year_cols]
```

```python
germany_co2_clean = germany_co2_clean.rename(columns={
    "Country Name": "country",
    "Country Code": "country_code"
})
```

```python
germany_co2_clean.columns
```

```python
germany_co2_clean.shape
```

```python
germany_co2_long = germany_co2_clean.melt(
    id_vars=["country", "country_code"],
    var_name="year",
    value_name="co2_per_capita"
)
```

```python
germany_co2_long["year"] = germany_co2_long["year"].astype(int)
```

```python
germany_co2_long = germany_co2_long.sort_values("year")
```

```python
germany_co2_long.shape
```

```python
germany_co2_long.head()
```

```python
germany_co2_long.isna().sum()
```

```python
germany_renewables_clean = germany_renewables[id_vars + year_cols]
```

```python
germany_renewables_clean = germany_renewables_clean.rename(columns={
    "Country Name": "country",
    "Country Code": "country_code"
})
```

```python
germany_renewables_clean.columns
```

```python
germany_renewables_long = germany_renewables_clean.melt(
    id_vars=["country", "country_code"],
    var_name="year",
    value_name="renewable_energy_share"
)
```

```python
germany_renewables_long["year"] = germany_renewables_long["year"].astype(int)
```

```python
germany_renewables_long.shape
```

```python
germany_renewables_long.head()
```

```python
germany_energy = germany_co2_long.merge(
    germany_renewables_long,
    on="year",
    how="inner"
)
```

```python
germany_energy.isna().sum()
```

```python
germany_energy = germany_energy.drop(
    columns=["country_y", "country_code_y"]
).rename(columns={
    "country_x": "country",
    "country_code_x": "country_code"
})
```

```python
germany_energy.columns
```

After merging CO₂ and renewable energy data, a single missing observation
was identified in the renewable energy series. Given the small magnitude,
the affected year was removed to ensure a complete and consistent dataset.

```python
germany_energy[germany_energy["renewable_energy_share"].isna()]
```

```python
germany_energy = germany_energy.dropna()
```

```python
germany_energy.head()
```

```python
plt.figure(figsize=(8,5))
plt.plot(
    germany_energy["year"],
    germany_energy["co2_per_capita"]
)
plt.xlabel("Year")
plt.ylabel("CO₂ emissions per capita (t)")
plt.title("Germany: CO₂ Emissions per Capita (1990-2020)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(
    germany_energy["year"],
    germany_energy["renewable_energy_share"]
)
plt.xlabel("Year")
plt.ylabel("Renewable energy share (%)")
plt.title("Germany: Renewable Energy Share (1990–2020)")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(germany_energy["year"], germany_energy["co2_per_capita"])
ax[0].set_title("CO₂ emissions per capita")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("t CO₂ per capita")

ax[1].plot(germany_energy["year"], germany_energy["renewable_energy_share"])
ax[1].set_title("Renewable energy share")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("% of total energy")

plt.tight_layout()
plt.show()
```

### Visual comparison (observations)

- Renewable energy share increases steadily over the full period, with faster growth after the mid-2000s.
- CO₂ emissions per capita show an overall decline, but with periods of stagnation.
- Periods of rapid renewable growth do not always coincide with equally rapid CO₂ reductions.


```python
germany_energy[["co2_per_capita", "renewable_energy_share"]].corr()
```

### Correlation analysis

A strong negative correlation (-0.91) is observed between renewable energy share
and CO₂ emissions per capita in Germany over the period 1990-2020.

This indicates that increases in renewable energy usage are associated with
lower per-capita CO₂ emissions. However, this relationship is descriptive and
does not establish causality. Other structural and policy factors likely
contribute to the observed trend.

### Limitations

- The analysis focuses on a single country and cannot be generalized.
- Correlation does not imply causation.
- The analysis does not control for economic growth, energy demand, or efficiency gains.

### Preliminary insights

- Germany’s increasing renewable energy share coincides with a long-term decline in per-capita CO₂ emissions.
- The relationship is not one-to-one, suggesting that factors such as industrial structure, energy demand, and efficiency also play a role.
- This highlights the importance of complementary climate and energy policies beyond renewable deployment alone.
